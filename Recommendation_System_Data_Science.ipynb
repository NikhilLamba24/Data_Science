````python
#Using Weighted average for each movie's Average Rating as per the formula
#W=(Rv+Cm)/v+m
# W= weighted rating
# R= avg for the movie as a no. from 0 to 10 = Rating
# v = no.of votes for the movies
# m= no. of votes to be listed in the Top 250( currently 3000)
# C= the mean vote across the whole report (currently 6.9)
````

````python
#Using Weighted average for each movie's Average Rating
v=movies_cleaned_df['vote_count']
R=movies_cleaned_df['vote_average']
C=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(0.70) # provided threshold here to be 70% means more than 70% votes should be there for a movie to be considered here
````

````python
movies_cleaned_df['weighted_average']=((R*v)+(C*m))/(v+m)
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3, max_features=None,
                      strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
                      ngram_range=(1,3),
                      stop_words='english')
#Filling NaNs with empty string
movies_cleaned_df['overview']= movies_cleaned_df['overview'].fillna('')
````

````python
#Fitting the TF-IDF on the 'overview' text
tfv_matrix = tfv.fit_transform(movies_cleaned_df['overview'])
````

````python
tfv_matrix
````

````python
tfv_matrix.shape
````

````python
from sklearn.metrics.pairwise import sigmoid_kernel

#compute the sigmoid kernel to get the output in between 0 and 1
sig=sigmoid_kernel(tfv_matrix, tfv_matrix)
````

````python
sig[0]
````

````python
indices = pd.Series(movies_cleaned_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
````

````python
def give_rec(title, sig=sig):
  # Get the index correspoingin to original_title
  idx = indices[title]

  #Get the pairwise similarity acores
  sig_scores = list(enumerate(sig[idx]))

  # sort teh movies
  sig_scores = sorted(sig_scores, key=lambda x:x[1], reverse=True)

  #scores of the 10 most similar movies
  sig_scores = sig_scores[1:11]

  #Movie indices
  movie_indices = [i[0] for i in sig_scores]

  #Top10 most similar movies
  return movies_cleaned_df['original_title'].iloc[movie_indices]
````
